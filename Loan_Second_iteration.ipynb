```python
import pandas as pd

mydata = pd.read_csv(r'sample_data/Loan.csv')
```

```python
import pandas as pd
from autogluon.tabular import TabularPredictor

# Load the trained predictor
predictor = TabularPredictor(label="LoanApproved", path="sample_data/loan_approval_model").fit(mydata)

# Evaluate on the same data or on a test dataset
evaluation_results = predictor.evaluate(mydata)
leaderboard = predictor.leaderboard(silent=True)

# Save evaluation results to a text file
with open("model_evaluation_results.txt", "w") as f:
    f.write("AutoGluon Model Evaluation Results\n")
    f.write("="*50 + "\n")
    for metric, value in evaluation_results.items():
        f.write(f"{metric}: {value}\n")

# Save leaderboard (ranking of all models) to CSV
leaderboard.to_csv("autogluon_leaderboard.csv", index=False)

print("Evaluation results saved to model_evaluation_results.txt")
print("Leaderboard saved to autogluon_leaderboard.csv")
```

```python
import pandas as pd
from autogluon.tabular import TabularPredictor

# Load the trained predictor
predictor = TabularPredictor(label="PreviousLoanDefaults", path="sample_data/loan_default_model").fit(mydata)


# Evaluate on the same data or on a test dataset
evaluation_results = predictor.evaluate(mydata)
leaderboard = predictor.leaderboard(silent=True)

output_path = "PreviousLoanDefaults_model_evaluation_results.txt"


# Save evaluation results to a text file
with open(output_path, "w") as f:
    f.write("AutoGluon Model Evaluation Results\n")
    f.write("="*50 + "\n")
    for metric, value in evaluation_results.items():
        f.write(f"{metric}: {value}\n")

output_path = "PreviousLoanDefaults_model_evaluation_results.txt"

# Save leaderboard (ranking of all models) to CSV
leaderboard.to_csv(output_path, index=False)
```
